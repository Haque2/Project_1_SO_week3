```python
average_income = soup.find_all('td', "fred-rls-elm-vl-td")[::3]
average_income_lists = [i.get_text().strip() for i in average_income]
average_income_lists
len(average_income_lists)
```

```python
name_state = soup.find_all('span', class_="fred-rls-elm-nm")
name_state_lists = [i.get_text() for i in name_state]
name_state_lists
len(name_state_lists)
```

```python
states = []
incomes = []

for i in name_state_lists:
    states.append(i)
    
for i in average_income_lists:
    incomes.append(i)
```

```python
income_state_df = pd.DataFrame({
    'State': states,
    'Av_income': incomes,
})
```

```python
income_state_df
```

```python
income_state_df = income_state_df.sort_values('State').reset_index(drop= True )
income_state_df
```

```python
income_state_df.to_csv('income_state.csv', index=False)
```
