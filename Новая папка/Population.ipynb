```python
name_state = soup.find_all("td", class_="z-40 border px-2 py-0.5")[::8]
name_state_lists = [i.get_text().strip() for i in name_state]

population = soup.find_all("td", class_="z-40 border px-2 py-0.5")[3::8]
population_lists = [i.get_text().strip() for i in population]

name_state = []
population = []

states = [i for i in name_state_lists]
population = [i for i in population_lists]



population_df = pd.DataFrame({'State': states, 'Population': population})
population_df
```

```python
population_df = population_df.sort_values('State').reset_index(drop= True )
population_df
```
